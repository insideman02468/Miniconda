```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
%pylab inline --no-import-all

map = Basemap()

map.drawcoastlines()

plt.show()
#plt.savefig('test.png')
```

```python
'''
All maps must have a projection. 
The projection and its features are all assigned when the object Basemap is created.
The way to do it is quite different from other libraries (i.e. GDAL),
so understanding this point is very important for working with Basemap.
'''

map = Basemap(projection='cyl')

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()

plt.show()
```

```python
'''
デフォルト値はcyl、またはEquirectangular projection
またはPlateCarréeとも呼ばれる円柱状の等距離投影です。
'''
map = Basemap(projection='aeqd', lon_0 = 10, lat_0 = 50)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()

plt.show()
```

```python
'''
この例では、UTM投影法を使用してメノルカ島、ゾーン31Nを示しています。
'''
map = Basemap(llcrnrlon=3.75,llcrnrlat=39.75,urcrnrlon=4.35,urcrnrlat=40.15, resolution = 'h', epsg=5520)

map.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()
plt.show()
```

```python
'''
拡張
今までの例はすべて地球全体を取ります。 領域のみを描画するには、バウンディングボックスを通過するか、マップの中心とマップサイズを渡します。
公式文書には、どちらの方法もほとんどの場合使用できると記載されていますが、多くの例外があります。
'''

map = Basemap(llcrnrlon=-10.5,llcrnrlat=35,urcrnrlon=4.,urcrnrlat=44.,
             resolution='i', projection='tmerc', lat_0 = 39.5, lon_0 = -3.25)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()

plt.show()
```
