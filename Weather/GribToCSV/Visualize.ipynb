```python
year="2019"
month="05"
day="10"
time="000000"
real_filename="realtime20190415000000.csv"

file_dir_path="Output_CSV/"+year+"/"+month+"/"+day

def getfile(year,month,day,time):
    file_name = file_dir_path+"/forecast"+str(year+month+day+time)+".csv"    
    return file_name

print(os.path.isfile( getfile(year,month,str(day),time) ))

def read_csv(year,month,day,time):
    input_data = pd.read_csv( getfile(year,month,day,time),encoding="UTF-8" ).set_index(['DateTime'])
    return input_data

def read_real_csv(real_filename):
    real_input="Realtime_CSV/"+real_filename
    return pd.read_csv( real_input,encoding="UTF-8" ).set_index(['Time'])
```

```python
data=pd.concat([read_csv(year,month,day,time),read_real_csv(real_filename)],axis=1,sort='False').dropna(axis = 0, how='any')
```

```python
data=data.loc[:, ['Date',
                  'wind_10m[m/s]', 'Real_Wind Speed(10m)[m/s]',
                'temperature','Real_Temperature(2m)[°C]', 
                'radiation','Real_Radiation[W/m2]',
                'Ppv(the output of the PV array)[Wh]','Real_Ppv(the output of the PV array)[Wh]'] ]
data
```

```python
from ipywidgets import interact
from matplotlib import cm, ticker
import matplotlib.dates as mdates
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

def figplot(size1,size2,fontsize,grid,y_range1,y_range2,y_range3,y_range4):
    rcParams['figure.figsize'] = size1,size2
    plt.rcParams["font.size"] = fontsize
    input1 = pd.to_datetime(data['Date'])
    fig = plt.figure(figsize=(50,30),dpi=200)
    fig = plt.figure()
    fig.set_tight_layout(False)
    

#####Wind
    real_wind = data['Real_Wind Speed(10m)[m/s]']
    forecast_wind = data['wind_10m[m/s]']
    
    ax1 = fig.add_subplot(2,2,1)
    #ax1.patch.set_facecolor('white')
    ax1.plot(input1, real_wind, color='Black',linestyle='solid', linewidth = 6, label='Realtime')
    ax1.set_ylim([0, y_range1])
    
    ax1.plot(input1,forecast_wind,linestyle='--', linewidth = 12, label="Forecast")

    ax1.set_title('Wind Speed(10m)[m/s]')
    ax1.set_xlabel('date')
    ax1.set_ylabel('Wind Speed(10m)[m/s]')
    ax1.set_xticklabels(input1,rotation=0, size="small")
    # 軸目盛の設定
    ax1.xaxis.set_major_locator(mdates.HourLocator(interval=grid))
    ax1.xaxis.set_major_formatter(mdates.DateFormatter("%dth %H"))
    plt.grid(which='major', color='black',linestyle='--')
    # 軸目盛ラベルの回転
    labels = ax1.get_xticklabels()
    plt.setp(labels, rotation=90);
    
    # 凡例をまとめて出力する
    fig.legend(loc="lower right",
               bbox_to_anchor=(1, 0.5, 0.4, .100), 
               borderaxespad=0.,
               ncol=1,
               mode="expand",
               title="LABEL NAME"
              ) 
    
#####Temp    
    real_temp = data['Real_Temperature(2m)[°C]']
    forecast_temp = data['temperature']
    
    ax2 = fig.add_subplot(2,2,2)
    #ax2.patch.set_facecolor('white')
    ax2.plot(input1, real_temp, color='Black',linestyle='solid', linewidth = 6)
    ax2.set_ylim([0, y_range2])
    
    ax2.plot(input1,forecast_temp,linestyle='--', linewidth = 12)

    ax2.set_title('Temperature[°C]')
    ax2.set_xlabel('date')
    ax2.set_ylabel('Temperature[°C]')
    ax2.set_xticklabels(input1,rotation=0, size="small")
    # 軸目盛の設定
    ax2.xaxis.set_major_locator(mdates.HourLocator(interval=grid))
    ax2.xaxis.set_major_formatter(mdates.DateFormatter("%dth %H"))
    plt.grid(which='major', color='black',linestyle='--')
    # 軸目盛ラベルの回転
    labels = ax2.get_xticklabels()
    plt.setp(labels, rotation=90);
    
#####radiation   
    real_rad = data['Real_Radiation[W/m2]']
    forecast_rad = data['radiation']
    
    ax3 = fig.add_subplot(2,2,3)
    #ax2.patch.set_facecolor('white')
    ax3.plot(input1, real_rad, color='Black',linestyle='solid', linewidth = 6)
    ax3.set_ylim([0, y_range3])
    
    ax3.plot(input1,forecast_rad,linestyle='--', linewidth = 12)

    ax3.set_title('Radiation[W/m2]')
    ax3.set_xlabel('date')
    ax3.set_ylabel('Radiation[W/m2]')
    ax3.set_xticklabels(input1,rotation=0, size="small")
    # 軸目盛の設定
    ax3.xaxis.set_major_locator(mdates.HourLocator(interval=grid))
    ax3.xaxis.set_major_formatter(mdates.DateFormatter("%dth %H"))
    plt.grid(which='major', color='black',linestyle='--')
    # 軸目盛ラベルの回転
    labels = ax3.get_xticklabels()
    plt.setp(labels, rotation=90);

#####Ppv(the output of the PV array)[Wh] 
    real_Ppv = data['Real_Ppv(the output of the PV array)[Wh]']
    forecast_Ppv = data['Ppv(the output of the PV array)[Wh]']
    
    ax4 = fig.add_subplot(2,2,4)
    #ax2.patch.set_facecolor('white')
    ax4.plot(input1, real_Ppv, color='Black',linestyle='solid', linewidth = 6)
    ax4.set_ylim([0, y_range4])
    
    ax4.plot(input1,forecast_Ppv,linestyle='--', linewidth = 12)

    ax4.set_title('Ppv(the output of the PV array)[Wh]')
    ax4.set_xlabel('date')
    ax4.set_ylabel('Ppv(the output of the PV array)[Wh]')
    ax4.set_xticklabels(input1,rotation=0, size="small")
    # 軸目盛の設定
    ax4.xaxis.set_major_locator(mdates.HourLocator(interval=grid))
    ax4.xaxis.set_major_formatter(mdates.DateFormatter("%dth %H"))
    plt.grid(which='major', color='black',linestyle='--')
    # 軸目盛ラベルの回転
    labels = ax4.get_xticklabels()
    plt.setp(labels, rotation=90);    
    
    plt.tight_layout()
    
```

```python
interact(figplot,size1=(0,130, 1),size2=(0,100, 1),fontsize=(0,100,1),grid=(1,12,1),y_range1=(1,25,1),y_range2=(1,60,1),\
        y_range3=(1,3000,1),y_range4=(1,500,1))
```
